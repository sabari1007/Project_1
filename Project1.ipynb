```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
%matplotlib inline

data = pd.read_csv('advertising.csv')
data.head()
```

```python
data.shape
```

```python
fig, axs = plt.subplots(1, 3, sharey=True, figsize=(16, 8))
data.plot(kind='scatter', x='TV', y='Sales', ax=axs[0])
data.plot(kind='scatter', x='Radio', y='Sales', ax=axs[1])
data.plot(kind='scatter', x='Newspaper', y='Sales', ax=axs[2])

axs[0].set_title('TV vs Sales')
axs[1].set_title('Radio vs Sales')
axs[2].set_title('Newspaper vs Sales')

plt.show()
```

```python
features_columns=['TV']
x=X=data[features_columns]
y=data.Sales

from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X, y)
print(lm.intercept_)
print(lm.coef_)
```

```python
add= lm.intercept_ + lm.coef_ * 50
print(add)
```
